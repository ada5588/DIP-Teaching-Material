#### Spatial Filters

```python
import cv2
import numpy as np
import scipy.ndimage
import matplotlib.pyplot as plt

# Opening the image using cv2.
a = cv2.imread('../Figures/airplane.png')  
#Ada:showing the original picture
a_ori = cv2.cvtColor(a, cv2.COLOR_BGR2RGB)
plt.title("original")
plt.imshow(a_ori)
plt.show()
# Converting the image to grayscale
a_gray = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
#Anto: before showing the pic,
#Anto: the color must be converted from BGR to RGB
#Anto: because openCV uses BGR while Matplotlib uses RGB
a_gray_show = cv2.cvtColor(a_gray, cv2.COLOR_BGR2RGB)
#Ada:showing gray scale picture
plt.title("gray scale")
plt.imshow(a_gray_show)
plt.show()
# Initializing the filter of size 5 by 5.
# The filter is divided by 25 for normalization.
k1 = np.ones((5,5))/25
# performing convolution
b = scipy.ndimage.filters.convolve(a_gray, k1)
b_show = cv2.cvtColor(b, cv2.COLOR_BGR2RGB)

plt.title("5 by 5 Spatial Filter output")
plt.imshow(b_show)
plt.show()
#Ada:try 10 by 10 filter
k2 = np.ones((10,10))/100
c = scipy.ndimage.filters.convolve(a_gray, k2)
c_show = cv2.cvtColor(c, cv2.COLOR_BGR2RGB)
plt.title("10 by 10 Spatial Filter output")
plt.imshow(c_show)
plt.show()
```

#### Median Filter

```python
import cv2
import scipy.ndimage
import matplotlib.pyplot as plt

# Opening the image and converting it to grayscale.
a = cv2.imread('../Figures/ct_saltandpepper.png')
a_show = cv2.cvtColor(a, cv2.COLOR_BGR2RGB)
plt.title("original")
plt.imshow(a_show)
plt.show()
#Ada: no need to convert the image to grayscale, 
#Ada: the original one is already grayscale
# a_gray = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
# Performing the size 5 median filter.
b = scipy.ndimage.filters.median_filter(a, size=5)
b_show = cv2.cvtColor(b, cv2.COLOR_BGR2RGB)
# Saving b as median_output.png in Figures folder
plt.title("size 5 median filter")
plt.imshow(b_show)
plt.show()
# Performing the size 14 median filter.
#Ada: as the size increases, it takes more time to finish processing
c = scipy.ndimage.filters.median_filter(a, size=14)
c_show = cv2.cvtColor(c, cv2.COLOR_BGR2RGB)
plt.title("size 14 median filter")
plt.imshow(c_show)
plt.show()
```

#### Sobel Filter

```python
import cv2
from scipy import ndimage
import matplotlib.pyplot as plt

# Opening the image.
a = cv2.imread('../Figures/cir.png')
# Converting a to grayscale .
a = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
#Ada:Converting a to RGB for showing
a_show = cv2.cvtColor(a, cv2.COLOR_BGR2RGB)
plt.imshow(a_show)
plt.title("original")
plt.show()
# Performing Sobel filter.
b = ndimage.sobel(a)
#Ada:Converting b to RGB for showing
b_show = cv2.cvtColor(b, cv2.COLOR_BGR2RGB)
plt.imshow(b_show)
plt.title("Sobel filter")
plt.show()
```

# Laplacian of Gaussian Filter

```python
import cv2
import scipy.ndimage
import matplotlib.pyplot as plt

# Opening the image.
a = cv2.imread('../Figures/vhuman_t1.png')
plt.title("original")
plt.imshow(a) 
plt.show()
# Performing Laplacian of Gaussian.
b = scipy.ndimage.filters.gaussian_laplace(a, sigma=1, mode='reflect')
plt.title("Laplacian of Gaussian Filter, sigma=1")
plt.imshow(b) 
plt.show()
#Ada: try sigma=1.5
c = scipy.ndimage.filters.gaussian_laplace(a, sigma=1.5, mode='reflect')
plt.title("Laplacian of Gaussian Filter, sigma=1.5")
plt.imshow(c) 
plt.show()
#Ada: try sigma=2
d = scipy.ndimage.filters.gaussian_laplace(a, sigma=2, mode='reflect')
plt.title("Laplacian of Gaussian Filter, sigma=2")
plt.imshow(d) 
plt.show()
```

# Frangi Filter

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

import numpy as np
from PIL import Image
from skimage.filters import frangi

img = cv2.imread('../Figures/angiogram1.png')
plt.title("original")
plt.imshow(img)
plt.show()
img1 = np.asarray(img)
img2 = frangi(img1, black_ridges=True)
img3 = 255*(img2-np.min(img2))/(np.max(img2)-np.min(img2))
plt.title("Frangi Filter, black_ridges=True")
plt.imshow(img3)
plt.show()
#Ada: try black_ridges=False
img4 = frangi(img1, black_ridges=False)
img5 = 255*(img4-np.min(img4))/(np.max(img4)-np.min(img4))
plt.title("Frangi Filter, black_ridges=False")
plt.imshow(img5)
plt.show()
```
