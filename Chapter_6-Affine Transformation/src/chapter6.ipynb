### Translation

```python
import numpy as np
import scipy.misc, math
from scipy.misc.pilutil import Image
from skimage.transform import AffineTransform, warp
import matplotlib.pyplot as plt

img = Image.open('../Figures/angiogram1.png')
img1 = scipy.misc.np.asarray(img)
plt.title("original")
plt.imshow(img1)
plt.show()
#Ada: translate by 100 pixels in x and 40 pixels in y
transformation1 = AffineTransform(translation=(100, 40))
img2 = warp(img1, transformation1)
plt.title("translation=(100, 40)")
plt.imshow(img2)
plt.show()
#Ada: translate by -40 pixels in x and -100 pixels in y
transformation2 = AffineTransform(translation=(-40, -100))
img3 = warp(img1, transformation2)
plt.title("translation=(-40, -100)")
plt.imshow(img3)
plt.show()
```

#### 6.2.2 Rotation

```python
import numpy as np
import scipy.misc,math
from scipy.misc.pilutil import Image
from skimage.transform import AffineTransform, warp
import matplotlib.pyplot as plt

img = Image.open('../Figures/angiogram1.png')
img1 = scipy.misc.np.asarray(img)
plt.title("original")
plt.imshow(img1)
plt.show()
# rotation angle in radians
transformation1 = AffineTransform(rotation=0.05)
img2 = warp(img1, transformation1)
plt.title("rotated by 0.05 radian")
plt.imshow(img2)
plt.show()
#Ada: rotation angle in radians
transformation2 = AffineTransform(rotation=0.8)
img3 = warp(img1, transformation2)
plt.title("rotated by 0.8 radian")
plt.imshow(img3)
plt.show()
#Ada: rotation angle in radians
transformation3 = AffineTransform(rotation=6.28)
img4 = warp(img1, transformation3)
plt.title("rotated by 6.28 radians")
plt.imshow(img4)
plt.show()
#Ada: rotation angle in radians
transformation4 = AffineTransform(rotation=-0.8)
img5 = warp(img1, transformation4)
plt.title("rotated by -0.8 radians")
plt.imshow(img5)
plt.show()
```

#### 6.2.3 Scaling

```python
import numpy as np
import scipy.misc, math
from scipy.misc.pilutil import Image
from skimage.transform import AffineTransform, warp
import matplotlib.pyplot as plt

img = Image.open('../Figures/angiogram1.png').convert('L')
img1 = scipy.misc.np.asarray(img)
plt.title("original")
plt.imshow(img1)
plt.show()
#scale by 1/2 on both x and y.
transformation1 = AffineTransform(scale= (0.5, 0.5))
img2 = warp(img1, transformation1)
plt.title("scale by 1/2 on both x and y.")
plt.imshow(img2)
plt.show()
#Ada: scale by 4 on x and by 1/3 on y.
transformation2 = AffineTransform(scale= (4, 1/3))
img3 = warp(img1, transformation2)
plt.title("scale by 4 on x and by 1/3 on y")
plt.imshow(img3)
plt.show()
```

#### 6.2.4 Interpolation

```python
import numpy as np
import scipy.misc, math
from scipy.misc.pilutil import Image
from skimage.transform import AffineTransform, warp
import matplotlib.pyplot as plt

img = Image.open('../Figures/angiogram1.png')
img1 = scipy.misc.np.asarray(img)

transformation = AffineTransform(scale=(0.07, 0.07))

# nearest neighbor order = 0
img2 = warp(img1, transformation, order=0)
plt.title("nearest neighbor")
plt.imshow(img2)
plt.show()

# bi-linear order = 1
img3 = warp(img1, transformation, order=1) 
plt.title("bi-linear")
plt.imshow(img3)
plt.show()

#bi-quadratic order = 2
img4 = warp(img1, transformation, order=2)
plt.title("bi-quadratic")
plt.imshow(img4)
plt.show()

#bi-cubic order = 3
img5 = warp(img1, transformation, order=3)
plt.title("bi-cubic")
plt.imshow(img5)
plt.show()
```
