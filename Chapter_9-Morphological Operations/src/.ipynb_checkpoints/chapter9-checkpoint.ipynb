#### Dilation

```python
from PIL import Image
import scipy.ndimage as snd
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Opening the image and converting it to grayscale.
a = Image.open('../figures/dil_image.png').convert('L')
a = np.array(a)
# Performing binary dilation for 3 iterations.
b = snd.morphology.binary_dilation(a, iterations=3)
#Ada: try 10 iterations
c = snd.morphology.binary_dilation(a, iterations=10)
#Ada: Compare results with original image
plt.title('Original')
imgplot = plt.imshow(a, cmap='gray',vmin=0,vmax=255)
plt.show()

plt.title('3 Iterations')
imgplot = plt.imshow(b, cmap='gray')
plt.show()
plt.title('10 Iterations')
imgplot = plt.imshow(c, cmap='gray')
plt.show()
```

#### Grayscale Dilation and Erosion

```python
import numpy as np
from PIL import Image
import scipy.ndimage
import matplotlib.pyplot as plt

# Opening the image and converting it into grayscale.
a = Image.open('../figures/sem3.png').convert('L')
# Creating a structuring element.
#Ada: try different sizes of footprint
footprint1 = np.ones((20 ,20))
footprint2 = np.ones((60 ,18))
footprint3 = np.ones((18 ,60))
# Performing grey dilation.
b = scipy.ndimage.morphology.grey_dilation(a, footprint=footprint1)
c = scipy.ndimage.morphology.grey_dilation(a, footprint=footprint2)
d = scipy.ndimage.morphology.grey_dilation(a, footprint=footprint3)
# Converting ndarray to image.
bi = Image.fromarray(b)
ci = Image.fromarray(c)
di = Image.fromarray(d)
#Ada:shouing the images
plt.title("original")
imgplot = plt.imshow(a, cmap='gray')
plt.show()
plt.title("(20 ,20)footprint")
imgplot = plt.imshow(bi, cmap='gray')
plt.show()
plt.title("(60 ,18)footprint")
imgplot = plt.imshow(ci, cmap='gray')
plt.show()
plt.title("(18 ,60)footprint")
imgplot = plt.imshow(di, cmap='gray')
plt.show()
```

#### Opening and Closing

```python
from PIL import Image
import scipy.ndimage as snd
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Opening the image and converting it to .
a = Image.open('../figures/dil_image.png').convert('L')
a = np.array(a)
# Defining the structuring element.
s = [[0,1,0],[1,1,1], [0,1,0]]
# Performing the binary opening for 5 iterations.
b = snd.morphology.binary_opening(a, structure=s, iterations=5)
#Ada: Performing the binary closing for 5 iterations.
c = snd.morphology.binary_closing(a,structure=s,iterations=5)
#Ada: compare the results
plt.title("Opening")
imgplot = plt.imshow(b, cmap='gray')
plt.show()

plt.title("original")
imgplot = plt.imshow(a, cmap='gray')
plt.show()

plt.title("Closing")
imgplot = plt.imshow(c, cmap='gray')
plt.show()
```

#### Grayscale Opening and Closing

```python

import numpy as np
from PIL import Image
import scipy.ndimage 
import matplotlib.pyplot as plt

# Opening the image and converting it into grayscale.
a = Image.open('../figures/adaptive_example1.png').convert('L')
# Creating a structuring element.
footprint = np.ones((40,40))
# Performing grey opening.
b = scipy.ndimage.morphology.grey_opening(a, footprint=footprint)
# Converting ndarray to image.
c = Image.fromarray(b)
# Saving the image.
c.save('../figures/grey_opening_output_40.png') 
fig = plt.figure()
ax = fig.add_subplot(1, 2, 1)
imgplot = plt.imshow(a, cmap='gray')
ax.set_title('Before')
ax = fig.add_subplot(1, 2, 2)
imgplot = plt.imshow(b, cmap='gray')
ax.set_title('After')
plt.show()
```

##### The Python code for grayscale closing is given below.

```python
import numpy as np
from PIL import Image
import scipy.ndimage 
import matplotlib.pyplot as plt

# Opening the image and converting it into grayscale.
a = Image.open('../figures/adaptive_example1.png').convert('L')
a = np.asarray(a)
# Creating a structuring element.
fp = np.ones((40,40))
# Performing grey closing.
bg = scipy.ndimage.morphology.grey_closing(a, footprint=fp)
# bg represents the background. 
# We will subtract bg from a to remove the background in a.
bg_free = (a.astype(np.float64) - bg.astype(np.float64))
# We rescale bg_free to 0 to 255.
denom = (bg_free.max()-bg_free.min())
bg_free_norm = (bg_free - bg_free.min())*255/denom
# Converting bg_free_norm to uint8.
bg_free_norm = bg_free_norm.astype(np.uint8)
# Converting bg_free_norm and bg to images.
bg_free_norm = Image.fromarray(bg_free_norm)
bg = Image.fromarray(bg)
# Saving the background image.
bg.save('../figures/grey_closing_out_40.png')  
# Saving the bg_free_norm image.
bg_free_norm.save('../figures/closing_bgfree.png') 
fig = plt.figure()
ax = fig.add_subplot(1, 2, 1)
imgplot = plt.imshow(a, cmap='gray')
ax.set_title('Before')
ax = fig.add_subplot(1, 2, 2)
imgplot = plt.imshow(b, cmap='gray')
ax.set_title('After')
plt.show()
```

### Skeletonization

```python
import numpy as np
from PIL import Image
from skimage.morphology import skeletonize
import cv2
import matplotlib.pyplot as plt

# Opening the image and converting it to grayscale.
a = Image.open('../figures//steps1.png').convert('L')
# Converting a to an ndarray and normalizing it.
a = np.asarray(a)/np.max(a)
# Performing skeletonization.
b = skeletonize(a)
# Saving the image as 8-bit as b is a
# binary image of dtype=bool
fig = plt.figure()
ax = fig.add_subplot(1, 2, 1)
imgplot = plt.imshow(a, cmap='gray')
ax.set_title('Before')
ax = fig.add_subplot(1, 2, 2)
imgplot = plt.imshow(b, cmap='gray')
ax.set_title('After')
plt.show()
```
